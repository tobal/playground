# Analyze IT salaries dataset

## Import data

```python
import pandas as pd
import numpy as np

salaries_df = pd.read_csv('salaries.csv')
```

## Munge data

```python
salaries_df['salary'] = salaries_df['salary'].str.replace(r'[^\d.]', '').map(lambda x: x[:-2]).astype('int')
```
