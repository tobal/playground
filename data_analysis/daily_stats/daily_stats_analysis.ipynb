```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

stats_df = pd.read_csv("stats.csv")
```

```python
stats_df.corr(method='pearson')
```

```python
stats_df.mean()
```

```python
def compute_and_plot_means(dataframe, column, aggr=10):
    means = pd.DataFrame()
    iters = len(dataframe) // aggr + 1
    for i in range(iters):
        df = pd.DataFrame(dataframe[column])[i * aggr:(i+1) * aggr]
        means[i] = df.mean()

    f = plt.figure(figsize=(20, 5))
    means.T.plot(ax=f.gca())
    plt.title('{} aggregated mean by {}'.format(column, aggr), color='black')
```

```python
columns = list(stats_df.columns)[1:]
for column in columns:
    for i in [2, 5, 10, 20, 30]:
        compute_and_plot_means(stats_df, column, i)
```
